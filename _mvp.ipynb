# Problem Definition
In the initial MVP,each card is represented by a binary state variable:
- 1 : correct
- 0 : Incorrect
Given a randomly generated binary sequence,the objective is to reach the all-ones state transition strategies

# Import

```python
#Standard Library
import random
import time
```

# Experiment Configuration

```python
Sequence_Length = 50
```

# Sequence Generator

```python
def generate_sequence(n):
    sequence = []
    for _ in range(1,n+1):
        sequence.append(random.randint(0,1))
    return sequence
sequence = generate_sequence(Sequence_Length)
print(sequence)
```

# Strategy A : Local Fix

```python
def local_fix(sequence):
    operation_times = 0
    for idx,state in enumerate(sequence):
        if state == 0:
            sequence[idx] = 1
            operation_times += 1
    return sequence,operation_times
local_test = sequence.copy()
print(local_fix(local_test))
```

# Strategy B : Suffix flip

```python
def Suffix_flip(sequence):
    operation_times = 0
    for idx,state in enumerate(sequence):
        if state == 0:
            for state_idx in range(idx,len(sequence)):
                sequence[state_idx] = 1 - sequence[state_idx]
            operation_times += 1
    return sequence,operation_times
Suffix_test = sequence.copy()
print(Suffix_flip(Suffix_test))
```

# Experiment part
Compare the two strategies on the same randomly generated binary sequence.

```python
# ==========================================================
# Step 1. Generate a random sequence
# ==========================================================
test_sequence = generate_sequence(Sequence_Length)

# ==========================================================
# Step 2. Run Local Fix Strategy
# ==========================================================
local_sequence = test_sequence.copy()
local_result,local_operation_times = local_fix(local_sequence)

# ==========================================================
# Step 3. Run Suffix Flip Strategy
# ==========================================================
Suffix_sequence = test_sequence.copy()
Suffix_result,Suffix_operation_times = Suffix_flip(Suffix_sequence)
# ==========================================================
# Step 4. Compare the Results
# ==========================================================
print("original:",test_sequence)
print("local result:",local_result)
print("local_times:",local_operation_times)
print("suffix result:",Suffix_result)
print("suffix_times:",Suffix_operation_times)
```
