## Monte Carlo Experiment
To eliminate randomness form a single trial,we repeat the experiment many times and compare the statistical performance of both strategies.
## Evaluation Metrics

This experiment evaluates each strategy using the following metrics:

1. Average number of operations (primary metric)
2. Average running time
3. Minimum number of operations
4. Maximum number of operations

## Import & Reuse Functions

```python
import random
from statistics import mean
# Generate sequence
def generate_sequence(n):
    sequence = []
    for _ in range(1,n+1):
        sequence.append(random.randint(0,1))
    return sequence
# Local_fix
def local_fix(sequence):
    operation_times = 0
    for idx,state in enumerate(sequence):
        if state == 0:
            sequence[idx] = 1
            operation_times += 1
    return sequence,operation_times
# Suffix_flip
def Suffix_flip(sequence):
    operation_times = 0
    for idx,state in enumerate(sequence):
        if state == 0:
            for state_idx in range(idx,len(sequence)):
                sequence[state_idx] = 1 - sequence[state_idx]
            operation_times += 1
    return sequence,operation_times
```

## Experiment Parameters

```python
Sequence_Length = 100
N_simulation = 10000
```

## Monte Carlo Simulation
Create two empty lists

↓

Repeat N_SIMULATIONS times

↓

Generate one random sequence

↓

Copy the sequence

↓

Run Strategy A

↓

Run Strategy B

↓

Record operation counts

↓

Repeat

```python
local_results = []
suffix_results = []
for experiment in range(N_simulation):
    sequence = generate_sequence(Sequence_Length)
    sequence_local = sequence.copy()
    sequence_suffix = sequence.copy()
    _,operations_local = local_fix(sequence_local)
    local_results.append(operations_local)
    _,operations_suffix = Suffix_flip(sequence_suffix)
    suffix_results.append(operations_suffix)
print("===== Strategy A =====")
print(f"Average Operations : {mean(local_results):.3f}")
print(f"Minimum Operations : {min(local_results)}")
print(f"Maximum Operations : {max(local_results)}")

print()

print("===== Strategy B =====")
print(f"Average Operations : {mean(suffix_results):.3f}")
print(f"Minimum Operations : {min(suffix_results)}")
print(f"Maximum Operations : {max(suffix_results)}")
```

## Findings

After conducting Monte Carlo simulations on uniformly random binary sequences,
the following observations were obtained:

- The average number of operations for both strategies converges to approximately half of the sequence length (n/2).
- No significant difference was observed between Local Fix and Suffix Flip under the current experimental setting.
- The experimental results remained consistent across different sequence lengths.
## Discussion

The initial hypothesis was that Suffix Flip might require fewer operations than Local Fix.
However, the Monte Carlo experiments did not support this hypothesis.

A possible explanation is that the uniformly random binary sequence preserves its statistical symmetry after each suffix flip, leading both strategies to have approximately the same expected number of operations.

A rigorous mathematical proof is left as future work.
